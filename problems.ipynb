```python
#problem 1

import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2, 100) #time

g = 9.81 #gravitational acceleration

#initial upward speed
v0_jim = 10
v0_mark = 15

y_jim = - 1/2 * g * t**2 + v0_jim * t
y_mark = - 1/2 * g * t**2 + v0_mark * t

plt.ylabel('Height')
plt.xlabel('Time')
plt.plot(t, y_jim, label = 'Jim')
plt.plot(t, y_mark, label = 'Mark')
plt.legend(fontsize = 12)
plt.grid()
plt.show()
```
